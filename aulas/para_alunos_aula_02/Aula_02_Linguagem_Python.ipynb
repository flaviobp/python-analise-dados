```python
# operacoes 
# uniao de dois conjuntos com a palavra chave union 
A = {0, 1, 3, 5, 7, 9} 
B = {0, 2, 4, 6, 8}
C = A.union(B)
print(C)
```

```python
# uniao de dois conjuntos com |  
A = {0, 1, 3, 5, 7, 9}
B = {0, 2, 4, 6, 8}
C = A | B  # union
print(C)
```

```python
# interseccao de dois conjuntos com a palavra chave intersect e com &
A = {0, 1, 3, 5, 7, 9}
B = {0, 2, 4, 6, 8}
C = A & B  # intersection C = A.intersect(B).

print(C)
```

## Funções 

* parametros:nomes dos atributos que serão passados para a função
* argumentos: são os valores passados para a função

```python
# como criar uma funcao que soma dois numeros?
def somar_numeros(x,y):
    return x+y
```

```python
# como chamar a funcao?
somar_numeros(1,2)
```

```python
# funcao com valores default
def somar_numeros(x=1,y=1):
    return x+y
```

```python
somar_numeros()
```

```python
# apenas um valor defaul
def somar_numeros(x,y=1): # atencao na ordem dos parametros default
    return x+y
somar_numeros(1)
```

```python
# atencao na ordem dos valores default 
def somar_numeros(y=1,x): # atencao na ordem dos parametros default
    return x+y
somar_numeros(1)
```

```python
# cuidado com as variaveis globais
def somar_numeros(x,y):
    return x+y+z

z = 2
somar_numeros(2,2)
```

```python
# sem o return, veja quais variáveis são impactadas
def func(x, y):
    x = x + 1
    y.append(4)

x = 1
y = [1, 2, 3]
func(x, y)
print('x:', x)
print('y:', y)
```

```python
# agora com return veja o impacto
def func(x, y):
    y.append(4)
    x = x + 1
    return x

x = 1
y = [1, 2, 3]
novo_x = func(x, y)
print('x:', x)
print('novo_x:',novo_x)
print('y:', y)
```

```python
# uso de args
def somar_numeros(*args):
    result = 0
    for i in args:
        result += i
    return result
```

```python
# chamar sem valores
print(somar_numeros())
# chamar com 2 valores
print(somar_numeros(1,4))
# chamar com 3 valores
print(somar_numeros(1,7,3))
```

```python
# kwargs
def print_args( **kwargs):
    print(kwargs)

print_args(param1=5, param2=6)
```

```python
# args e kwargs
def print_args(a, b, *args,  **kwargs): # args precisa estar primeiro
    print(a, b)
    print(args)
    print(kwargs)
    
print_args(1, 4, 6, 5, param1=5, param2=6)
```
